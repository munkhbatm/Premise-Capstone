## Create a lag data

```python
df2 = data.copy()
```

```python
df2['Lag_Date'] = df2['Last_Update'] + timedelta(days=1)
```

```python
df2
```

```python
df2.rename(columns={'Confirmed':'Confirmed_pre', 'Deaths':'Deaths_pre','Recovered':'Recovered_pre','Active':'Active_pre'}, inplace=True)
df2.drop(columns=['Last_Update'], inplace=True)
```

```python
df2
```

```python
data=pd.merge(data, df2, left_on=['Country_Region','Last_Update'], right_on=['Country_Region','Lag_Date'], how='left')
```

```python
data['Confirmed_new'] = data['Confirmed'] - data['Confirmed_pre']
data['Deaths_new'] = data['Deaths'] - data['Deaths_pre']
data['Recovered_new'] = data['Recovered'] - data['Recovered_pre']
data['Active_new'] = data['Active'] - data['Active_pre']
```

```python
data.drop(columns=['Confirmed_pre','Deaths_pre','Recovered_pre','Active_pre','Lag_Date'], inplace=True)
```

## rename countries

```python
def country_recode(series):
    if series == "Cote d'Ivoire":
        return  "Côte d'Ivoire"
    if series == "Congo (Kinshasa)":
        return "Democratic Republic of the Congo"
    if series == "Burma":
        return "Myanmar"
    if series == "Congo (Brazzaville)":
        return "Republic of Congo"
    if series == "Korea, South":
        return "South Korea"
    if series == "Taiwan*":
        return "Taiwan"
    if series == "US":
        return "United States" 
    else:
        return series
      
data_week['Country_Region'] = data_week['Country_Region'].apply(country_recode)
```
